```python
# Paper folding problem
# time to run
import time
start = time.time()

THICKNESS = 0.00008
FOLD = 1

# creating a list
thickness_list = []
thickness_list.append(THICKNESS)
folds_list = []
folds_list.append(FOLD)

#start with initial thickness
folded_thickness = THICKNESS

#fold the paper 43 times, doubling thickness each time

for fold in range (43):
  folded_thickness = folded_thickness * 2
  new_thickness = thickness_list[-1] * 2 #doubles the previous thickness
  thickness_list.append(new_thickness)
  new_fold = folds_list[-1] + 1 # adds one to the folds list
  folds_list.append(new_fold)

print(folds_list) # prints the final folds list
print(thickness_list) # prints the final thickness list

if len(thickness_list) == 44:
  print("The list contains 44 values")
else:
  print("The list does not contain 44 values")

#convert to kilimetres and disply with 2 decimal points
print("Thickness after 43 folds: {:.2f} kilometres".format(folded_thickness/1000))

# uses the folds list and thickness list to plot a graph
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(folds_list, thickness_list)
plt.title("Thickness of folded paper")
plt.xlabel("Number of folds")
plt.ylabel("Thickness [m]")
plt.show()

#graph 2
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(folds_list, thickness_list, color = 'red', label = 'Red line')
plt.title("Thickness of folded paper with a red line")
plt.xlabel("Number of folds")
plt.ylabel("Thickness [m]")
plt.show()

#graph 3
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(folds_list, thickness_list, linewidth = 3, label = 'Thick line')
plt.title("Thickness of folded paper with a thickened line")
plt.xlabel("Number of folds")
plt.ylabel("Thickness [m]")
plt.show()

#graph 4
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(folds_list, thickness_list, linestyle = ':', color = 'blue', label = 'Dotted line')
plt.title("Thickness of folded paper with a dotted line")
plt.xlabel("Number of folds")
plt.ylabel("Thickness [m]")
plt.show()


# time to run
elapsed_time = time.time() - start
print("time : {} [s]".format(elapsed_time))
```
